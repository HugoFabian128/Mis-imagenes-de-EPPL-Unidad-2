En el estudio de la geometría, es fundamental comprender las propiedades de las figuras circulares. El círculo es una figura geométrica plana formada por todos los puntos que se encuentran a una distancia constante llamada radio desde un punto central.
Este programa tiene como objetivo desarrollar una aplicación interactiva que permita calcular las principales medidas del círculo y la esfera mediante un menú de opciones. El usuario podrá seleccionar entre calcular el área del círculo (superficie que ocupa), el perímetro o circunferencia (longitud del borde), o el volumen de una esfera (figura tridimensional generada al rotar un círculo).
La aplicación utilizará funciones independientes para cada cálculo, facilitando la organización del código y su reutilización, implementando estructuras de control condicionales y cíclicas para crear una experiencia de usuario amigable e intuitiva.
1. Área del círculo = π * r2
2. erímetro del círculo = P =2 × π × r
3. Volumen de la esfera = = 3/4 × π × r 3

```python
🎯 CARACTERÍSTICAS DEL PROGRAMA
✅ Estructura modular con 4 funciones independientes
✅ Menú interactivo con emojis y diseño elegante
✅ Uso de estructuras de control: if, elif, else, while
✅ Llamadas a funciones desde el menú principal
✅ Solicitud de radio dentro de cada función
✅ Fórmulas matemáticas precisas usando math.pi
✅ Resultados formateados con 4 decimales
✅ Sin programación orientada a objetos
✅ Código en un solo archivo listo para ejecutar
```

```python
import math
# ═══════════════════════════════════════════════════════════
# FUNCIÓN: areacirculo()
# Calcula el área de un círculo dado su radio
# Fórmula: A = π × r²
# FUNCIÓN: perimetrocirculo()
# Calcula el perímetro (circunferencia) de un círculo
# Fórmula: P = 2 × π × r
# FUNCIÓN: volumenesfera()
# Calcula el volumen de una esfera dado su radio
# Fórmula: V = (4/3) × π × r³
# FUNCIÓN: menu()
# Muestra el menú de opciones y retorna la selección
# PROGRAMA PRINCIPAL
# ═══════════════════════════════════════════════════════════
def areacirculo():
    print("\n📐 CÁLCULO DEL ÁREA DEL CÍRCULO")
    print("─" * 40)
    radio = float(input("Ingrese el valor del radio: "))
    area = math.pi * (radio ** 2)
    print(f"\n✅ Resultados:")
    print(f"   Radio: {radio}")
    print(f"   Área: {area:.4f} unidades cuadradas")
    print("─" * 40)


def perimetrocirculo():
    print("\n📏 CÁLCULO DEL PERÍMETRO DEL CÍRCULO")
    print("─" * 40)
    radio = float(input("Ingrese el valor del radio: "))
    perimetro = 2 * math.pi * radio
    print(f"\n✅ Resultados:")
    print(f"   Radio: {radio}")
    print(f"   Perímetro: {perimetro:.4f} unidades")
    print("─" * 40)


def volumenesfera():
    print("\n🔮 CÁLCULO DEL VOLUMEN DE LA ESFERA")
    print("─" * 40)
    radio = float(input("Ingrese el valor del radio: "))
    volumen = (4/3) * math.pi * (radio ** 3)
    print(f"\n✅ Resultados:")
    print(f"   Radio: {radio}")
    print(f"   Volumen: {volumen:.4f} unidades cúbicas")
    print("─" * 40)


def menu():
    print("\n" + "╔" + "═" * 48 + "╗")
    print("║" + " " * 10 + "🔵 MENÚ PRINCIPAL 🔵" + " " * 16 + "║")
    print("╠" + "═" * 48 + "╣")
    print("║  1. Calcular área del círculo           ⭕  ║")
    print("║  2. Calcular perímetro del círculo      📏  ║")
    print("║  3. Calcular volumen de la esfera       🔮  ║")
    print("║  4. Salir del programa                  👋  ║")
    print("╚" + "═" * 48 + "╝")
    opcion = input("\n💡 Seleccione una opción (1-4): ")
    return opcion

print("\n" + "" * 25)
print("   BIENVENIDO AL SISTEMA DE CÁLCULO GEOMÉTRICO")
print("🌟" * 25)

opcion = 0

while opcion != "4":
    opcion = menu()

    if opcion == "1":
        areacirculo()
    elif opcion == "2":
        perimetrocirculo()
    elif opcion == "3":
        volumenesfera()
    elif opcion == "4":
        print("\n" + "✨" * 40)
        print("   🎉 ¡Gracias por usar el programa! 🎉")
        print("   👨‍ Desarrollado para clase de Programación")
        print("✨" * 40 + "\n")
    else:
        print("\n⚠️  Opción no válida. Por favor, seleccione 1, 2, 3 o 4.")

    if opcion != "4":
        input("\n⏸️  Presione ENTER para continuar...")

print("✅ Programa finalizado exitosamente.\n")
```
